# <font color=#3D6BBC > 1. Le format CSV</font>

Le format csv (pour comma separated values, soit en français valeurs séparées par des virgules) est un format très pratique pour représenter des données structurées.     
Dans ce format, chaque ligne représente un <font size='3' color=#3D6BBC><strong> <em>enregistrement</em></strong> </font>  et, sur une même ligne, les différents <font size='3' color=#3D6BBC><strong> <em>champs </em></strong> </font> de l’enregistrement sont réparés par une virgule (d’où le nom).   
En pratique, on peut spécifier le caractère utilisé pour séparer les différents champs et on utilise fréquemment un point-virgule, une tabulation ou deux points pour cela.   
<strong>Notons enfin que la première ligne d’un tel fichier est souvent utilisée pour indiquer le nom des différents champs. Dans ce cas, le premier enregistrement apparaîssant en deuxième ligne du fichier. </strong>

# <font color=#3D6BBC > 2. Importation des données vers une matrice (tableau de tableau)</font>

Une façon de charger un fichier csv en Python est d’utiliser la bibliothèque du même nom.   
Voici une portion de code permettant de charger le fichier prenoms.csv avec des points-virgules comme délimitations.

```python
import csv

#ouverture du fichier csv
with open('prenoms.csv', newline='') as csvfile:
    file = csv.reader(csvfile, delimiter=';')#lecture du fichier
    #création du tableau prenoms
    prenoms = []
    for row in file:
        prenoms.append(row)
```

Dans ce cas, les résultats sont stockés sous forme d’un tableau de tableau aussi appelé __matrice__. On a par exemple :

```python
import numpy as np
print(np.array(prenoms))
```

On a par exemple :

```python
prenoms[0]
```

```python
prenoms[5]
```

```python
prenoms[5][3]  # 5 étant le numéro de ligne et 3 le numéro de la colonne
```

# <font color=#3D6BBC > 3. Importation des données vers un tableau de p-uplets nommés </font>

Pour y remédier, nous allons utiliser une alternative offerte par le module __csv__:la fonction __DictReader__.   
Cette fonction renvoie un dictionnaire pour chaque enregistrement, la première ligne (du fichier csv) étant utilisée pour nommer les différents champs.  

```python
import csv
import numpy as np
prenoms_dict=[]
with open('prenoms.csv', newline='') as csvfile:
    file = csv.DictReader(csvfile, delimiter=';')
    for row in file:
        prenoms_dict.append(dict(row))
print(np.array(prenoms_dict))
```

Cette fois, nous obtenons un tableau de __p-uplets__ représentés sous forme de dictionnaires : <font color=red > les p_uplets nommés </font>

## <font color=#3D6BBC > Les p-uplets nommés</font>

Les p-uplets nommés sont des p-uplets mais on accède aux éléments grâce à un descripteur, et pas  par leur indice.  
__Ce type n'existe pas nativement en Python__.   
Le principal intérêt de ce type est d’améliorer la lisibilité du code, et donc de réduire les risque d’erreurs.

Les p-uplets nommés sont normalement __immutables__ alors que les dictionnaires (que nous utiliserons en Python pour les
représenter) ne le sont pas. Ainsi, dans la suite, nous nous garderons de modifier les dictionnaires représentant des p-uplets.

```python
print(prenoms_dict[0])
```

Ainsi, la 1ère ligne ne contient plus le noms des champs mais le 1er enregistrement.

## <font color=#3D6BBC > 4.2 Tri des données selon l'occurence croissante avec la fonction sorted</font>

![center](fonction-sorted.png)

La méthode __sort__ trie la liste en place, alors que la fonction __sorted__ renvoie une nouvelle liste correspondant la liste triée, la liste initiale étant laissée intacte.

__<u>1ère étape</u>__ :  créer une fonction cle_occurence ayant pour argument une ligne et qui renvoie la valeur de la __clé 'Occurence'__ dans chaque ligne (p-uplet nommé) de la table:

```python
def cle_occurence(p):
    return float(p['Occurence'])
```

__<u>2ème étape</u>__ : passer cette fonction en paramètre key dans la focntion sorted.

```python
prenoms_trier=sorted(prenoms_dict, key=cle_occurence)
print(np.array(prenoms_trier))
```

## <font color=#3D6BBC > 4.3 Tri des données selon plusieurs critères</font>

On souhaite trier la table prenoms_dict selon deux critères :les prénoms par ordre alphabétiques  puis l'occurence croissante.

```python
#création de la fonction qui renvoie une paire de clé(prénom, occurence)
def cle_alpha_occurence(p):
    return (p['Prénom'],float(p['Occurence']))
```

# <font color=#3D6BBC > 5. Fusion de tables</font>

Nous disposons maintenant de deux fichiers csv personnes_1.csv et personnes_2.csv. 

<u>__personnes_1.csv__ :</u> ![center](personnes_1.png)

<u>__personnes_2.csv__ :</u> ![center](personnes_2.png)

Comme vous pouvez le voir ci-dessus, les données sont séparées par des virgules et non par des points-virgules.

<u> __Question 6:__ </u>  Ecrire le code permettant de créer deux tables (personnes_1 et personnes_2) à partir de ces deux fichiers. 

```python
import csv
import numpy as np
personnes_1=[]
personnes_2=[]
with open('personnes_1.csv', newline='') as csvfile:
    file = csv.DictReader(csvfile, delimiter=',')
    for row in file:
        personnes_1.append(dict(row))
print(np.array(personnes_1))
```

```python
with open('personnes_2.csv', newline='') as csvfile:
    file = csv.DictReader(csvfile, delimiter=',')
    for row in file:
        personnes_2.append(dict(row))
print(np.array(personnes_2))
```

Pour fusionner ces deux tables, il suffit de faire:

```python
personnes=personnes_1 + personnes_2  #on concatène les deux tables
print(np.array(personnes))
```

<u> __Question 7:__ </u>  Ecrire le code permettant de trier les personnes par âge croissant.

```python
def cle_age(p):
    return (float(p['age']))
personnes_trier=sorted(personnes, key=cle_age)
print(np.array(personnes_trier))
```
